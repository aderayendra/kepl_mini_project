## 1. Get Mahasiswa

**Endpoint**
```
GET /api/mahasiswa
```

**Description**
Retrieve mahasiswa information.

### Example Request

````
import requests

url = "http://localhost:5000/api/mahasiswa"
response = requests.get(url)
response.json()
````

### Success Response (200 OK)
```json
{
  "count": 2,
  "mahasiswa": [
    {
      "nim": 12345678,
      "nama": "Budi Santoso",
      ...
    },
    ...
  ]
}

```

### Empty Response
```json
{
  "count": 0,
  "mahasiswa": []
}
```

## 2. Mahasiswa Login

**Endpoint**
```
POST /api/mahasiswa/login
```

**Description**
Authenticate mahasiswa using NIM and password.

**Headers**
```
Content-Type: application/json
```

### Example Request

````
import requests

url = "http://localhost:5000/api/mahasiswa/login"
payload = {
    "nim": 12345678,
    "password": "password123"
}

response = requests.post(url, json=payload)
response.json()
````

### Success Response (200 OK)
```json
{
  "authenticated": true,
  "mahasiswa": {
    "nim": 12345678,
    "nama": "Budi Santoso",
    ...
  }
}
```
